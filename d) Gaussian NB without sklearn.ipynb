```python
# after removing data row as index row and changing column names
import pandas as pd
import numpy as np
df=pd.read_csv('C:\\Users\\soham\\Downloads\\adult.data',names=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income'],header=None)
df 
```

```python
#replacing "?" by np.nan
df=df.replace(' ?',np.nan)
#replace missing values by mean,mode
for i in df.columns:
    if (type(df[i].dropna().tolist()[0])==str): #we are checking type of first value of each column after dropping NA's from it.
        df[i]=df[i].fillna(df[i].mode()[0])   #So, no problem even if first value of column turns out to be missing
    elif (type(df[i].dropna().tolist()[0])==float):  #we are checking type of first available element of that column using dropna 
        df[i]=df[i].fillna(df[i].mean())
df
```

```python
#creating a dataframe with only integer/numerical type columns
dfint=pd.DataFrame()
for i in df.columns:
    if (type(df[i].tolist()[0])==int):
        dfint[i]=df[i].tolist()
dfint
```

```python
#performing train-test-split on our dataframe
# X consists of only integer type columns, y consists of output column("income")
import numpy as np
from sklearn.model_selection import train_test_split
X, y = dfint.iloc[:,:],df.iloc[:,[-1]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
```

```python
#to obtain a dictionary of the class priors of output column
def getClassPriors(opcol):
    opcol=list(opcol)
    CP={i:opcol.count(i)/len(opcol) for i in np.unique(opcol)}
    return(CP)

getClassPriors(y_train['income'])
```

```python
#mean of each column corresponding to each output value
def getmeanval(X_train,y_train,opcol):
    mean={i:[X_train[j][(y_train[opcol]==i)].mean() for j in X_train.columns] for i in np.unique(y_train[opcol]) }
    return mean

getmeanval(X_train,y_train,'income')
```

We can directly use .std() function (just like we used .mean() above) instead of (.var()**0.5) but .std() does include smoothing coefficient (max variance * 10^-9) and also it takes ddof=N-1 by default, so we need to change ddof=0 and add smoothing coefficient or else results & final accuracy won't be the same as that of sklearn gaussian NB.

```python
#mean of each column corresponding to each output value
def getstdval(X_train,y_train,opcol):
    var_dict={i:[(X_train[j][(y_train[opcol]==i)].var(ddof=0)) for j in X_train.columns ] for i in np.unique(y_train[opcol]) }
    #find the max variance (for every possible output value) and add (max(var)/10**9) to original variance while calculating std for smoothing purposes
    #ddof=0 implies we would divide by 'N' while calculating variance, not by 'N-1'
    std_dict={i:[(X_train[j][(y_train[opcol]==i)].var(ddof=0)+ (max(var_dict[i])/10**9))**0.5  for j in X_train.columns ] for i in np.unique(y_train[opcol]) }
    return std_dict

getstdval(X_train,y_train,'income')
```

```python
#calculating the conditional probability of a given row for a given output value
import math
def condprob(x,mean,std,classpriors,opval):
    num = [math.exp(-(float(i)-float(j))**2/(2*(float(k)**2))) for i,j,k in zip(x,mean[opval],std[opval]) ]
    denom = [(2*math.pi*(float(k)**2))**0.5 for k in std[opval]]
    l=[(a/b) for a,b in zip(num,denom) ]
    condprobability=np.prod(l)*classpriors[opval]
    return condprobability

condprob(X_test.values[0],getmeanval(X_train,y_train,'income'),getstdval(X_train,y_train,'income'),getClassPriors(y_train['income']),' >50K')
```

```python
# finding the prediction (<50K or >=50K) for a single row
def predictSingle(x,mean,std,classpriors):
    predvals={opval:condprob(x,mean,std,classpriors,opval) for opval in classpriors.keys() }
    print(predvals)
    return max(predvals, key=predvals.get)

predictSingle(X_test.values[0],getmeanval(X_train,y_train,'income'),getstdval(X_train,y_train,'income'),getClassPriors(y_train['income']))
```

```python
# finding the prediction (<50K or >=50K) of all the rows and hence finding the accuracy of our prediction by comparing it with actual values
def predictMultiple(X_test,mean,std,classpriors):
    predlist=X_test.apply(lambda row:predictSingle(row.tolist(),mean,std,classpriors), axis=1 )
    return predlist
    
predictMultiple(X_test,getmeanval(X_train,y_train,'income'),getstdval(X_train,y_train,'income'),getClassPriors(y_train['income']))
```

```python
def accuracy(actlist,predlist):
    acc=(sum([a==b for (a,b) in zip(predlist,actlist)])/len(predlist))*100
    print('Accuracy of Algorithm is : ',acc,'%')
    
accuracy(y_test['income'].tolist(),predictMultiple(X_test,getmeanval(X_train,y_train,'income'),getstdval(X_train,y_train,'income'),getClassPriors(y_train['income'])))
```
