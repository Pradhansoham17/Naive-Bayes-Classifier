```python
# after removing data row as index row and changing column names
import pandas as pd
import numpy as np
df=pd.read_csv('C:\\Users\\soham\\Downloads\\adult.data',names=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income'],header=None)
df  
```

```python
#replacing "?" by np.nan
df=df.replace(' ?',np.nan)
#replace missing values by mean,mode
for i in df.columns:
    if (type(df[i].dropna().tolist()[0])==str): #we are checking type of first value of each column after dropping NA's from it.
        df[i]=df[i].fillna(df[i].mode()[0])   #So, no problem even if first value of column turns out to be missing
    elif (type(df[i].dropna().tolist()[0])==float):  #we are checking type of first available element of that column using dropna 
        df[i]=df[i].fillna(df[i].mean())
df
```

```python
#creating a dataframe with only integer/numerical type columns
dfint=pd.DataFrame()
for i in df.columns:
    if (type(df[i].tolist()[0])==int):
        dfint[i]=df[i].tolist()
dfint
```

```python
#creating a dataframe with only integer/numerical type columns
dfobj=pd.DataFrame()
for i in df.columns:
    if (type(df[i].tolist()[0])==str):
        dfobj[i]=df[i].tolist()
dfobj
```

```python
#performing train-test-split on our dataframe
import numpy as np
from sklearn.model_selection import train_test_split

# X consists of only categorical columns, y consists of output column("income")
X, y = dfobj.iloc[:,0:7],dfobj.iloc[:,[-1]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)

# X1 consists of only integer type columns, y1 consists of output column("income")
X1, y1 = dfint.iloc[:,:],df.iloc[:,[-1]]
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.3, random_state=5)

# X2 consists of all input columns, y2 consists of output column("income")
X2, y2 = df.iloc[:,0:13],df.iloc[:,[-1]]
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.3, random_state=5)
```

```python
#to obtain a dictionary of the class priors of output column
def getClassPriors(opcol):
    opcol=list(opcol)
    CP={i:opcol.count(i)/len(opcol) for i in np.unique(opcol)}
    return(CP)

getClassPriors(y_train['income'])
```

```python
# calculating the conditional probability of each category for each output value
def getcondProbs_obj(inp_traindf,op_traindf,opcol):
    Conpro={j:{k:(  len(inp_traindf[i][(inp_traindf[i]==j) & (op_traindf[opcol]==k)])  /  len(inp_traindf[i])  )   /   (  len(op_traindf[opcol][(op_traindf[opcol]==k)])  /  len(op_traindf[opcol])  )  for k in np.unique(op_traindf[opcol]) }  for i in inp_traindf.columns.values  for j in np.unique(inp_traindf[i]) }               
    return Conpro 

getcondProbs_obj(X_train,y_train,'income')       
```

```python
#mean of each column corresponding to each output value
def getmeanval_int(X_train,y_train,opcol):
    CP={i:[X_train[j][(y_train[opcol]==i)].mean() for j in X_train.columns] for i in np.unique(y_train[opcol]) }
    return CP

getmeanval_int(X1_train,y1_train,'income')
```

```python
#mean of each column corresponding to each output value
def getstdval_int(X_train,y_train,opcol):
    var_dict={i:[(X_train[j][(y_train[opcol]==i)].var(ddof=0)) for j in X_train.columns ] for i in np.unique(y_train[opcol]) }
    #find the max variance (for every possible output value) and add (max(var)/10**9) to original variance while calculating std for smoothing purposes
    #ddof=0 implies we would divide by 'N' while calculating variance, not by 'N-1'
    std_dict={i:[(X_train[j][(y_train[opcol]==i)].var(ddof=0)+ (max(var_dict[i])/10**9))**0.5  for j in X_train.columns ] for i in np.unique(y_train[opcol]) }
    return std_dict

getstdval_int(X1_train,y1_train,'income')
```

```python
#calculating the conditional probability of a given row for a given output value
import math
def condprob_int(x,mean,std,classpriors,opval):
    num = [math.exp(-(float(i)-float(j))**2/(2*(float(k)**2))) for i,j,k in zip(x,mean[opval],std[opval]) ]
    denom = [(2*math.pi*(float(k)**2))**0.5 for k in std[opval]]
    l=[(a/b) for a,b in zip(num,denom) ]
    condprobability=np.prod(l)*classpriors[opval]
    return condprobability
#X1_test.values[0]
condprob_int(X1_test.iloc[0,0:6],getmeanval_int(X1_train,y1_train,'income'),getstdval_int(X1_train,y1_train,'income'),getClassPriors(y1_train['income']),' >50K')
```

```python
# predicting the output of a single row 
# for each possible output value, (multiplying together the condprob of each element of input row)
# and then declaring the greater value of multiplication as our predection

def predictone(x_obj,x_int,mean,std,condprobs,classpriors):
    outprob={}
    for i in classpriors.keys():
        l=[condprobs[x_obj[j]][i] for j in x_obj.index]
        outprob[i]=np.prod(l)*classpriors[i]
    print(outprob)  
    predvals={opval:condprob_int(x_int,mean,std,classpriors,opval) for opval in classpriors.keys() }
    print(predvals)
    print(max(outprob, key=outprob.get)) 
    print(max(predvals, key=predvals.get))
    #comb={a:b*c for a in predvals.keys() for b,c in zip(outprob.values(),predvals.values())}
    #print(comb)

predictone(X_test.iloc[0,0:7], X1_test.iloc[0,0:6], getmeanval_int(X1_train,y1_train,'income'), getstdval_int(X1_train,y1_train,'income'),  getcondProbs_obj(X_train,y_train,'income'),  getClassPriors(y_train['income'])) 
```

```python
def predictall(X_test,X1_test,X2_test,mean,std,condprobs,classpriors):
    y_pred=X2_test.apply(lambda row:predictone(row[X_test.columns],row[X1_test.columns],mean,std,condprobs,classpriors), axis=1 )
    return y_pred

predictall(X_test,X1_test,X2_test,getmeanval_int(X1_train,y1_train,'income'), getstdval_int(X1_train,y1_train,'income'),  getcondProbs_obj(X_train,y_train,'income'),  getClassPriors(y_train['income']))   
```

```python
def accuracy(actlist,predlist):
    acc=(sum([a==b for (a,b) in zip(predlist,actlist)])/len(predlist))*100
    print('Accuracy of Algorithm is : ',acc,'%')
    
accuracy(y_test['income'].tolist(), predictall(X_test,X1_test,X2_test,getmeanval_int(X1_train,y1_train,'income'), getstdval_int(X1_train,y1_train,'income'),  getcondProbs_obj(X_train,y_train,'income'),  getClassPriors(y_train['income'])))
```
