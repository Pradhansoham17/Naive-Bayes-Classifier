```python
# after removing data row as index row and changing column names
import pandas as pd
import numpy as np
df=pd.read_csv('C:\\Users\\soham\\Downloads\\adult.data',names=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','income'],header=None)
df  
```

```python
#replacing "?" by np.nan
df=df.replace(' ?',np.nan)
#replace missing values by mean,mode
for i in df.columns:
    if (type(df[i].dropna().tolist()[0])==str): #we are checking type of first value of each column after dropping NA's from it.
        df[i]=df[i].fillna(df[i].mode()[0])   #So, no problem even if first value of column turns out to be missing
    elif (type(df[i].dropna().tolist()[0])==float):  #we are checking type of first available element of that column using dropna 
        df[i]=df[i].fillna(df[i].mean())
df
```

```python
#creating a dataframe with only integer/numerical type columns
dfint=pd.DataFrame()
for i in df.columns:
    if (type(df[i].tolist()[0])==int):
        dfint[i]=df[i].tolist()
dfint
```

```python
#creating a dataframe with only integer/numerical type columns
dfobj=pd.DataFrame()
for i in df.columns:
    if (type(df[i].tolist()[0])==str):
        dfobj[i]=df[i].tolist()
dfobj
```

```python
#performing train-test-split on our dataframe
import numpy as np
from sklearn.model_selection import train_test_split

# X consists of only categorical columns, y consists of output column("income")
X, y = dfobj.iloc[:,0:7],dfobj.iloc[:,[-1]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)

# X1 consists of only integer type columns, y1 consists of output column("income")
X1, y1 = dfint.iloc[:,:],df.iloc[:,[-1]]
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.3, random_state=5)

# X2 consists of all input columns, y2 consists of output column("income")
X2, y2 = df.iloc[:,0:13],df.iloc[:,[-1]]
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.3, random_state=5)
```

```python
import numpy as np
X1 = np.array(X1_train.values)
Y1 = np.array(y1_train.values)
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X1, Y1)

#conditional probability outcome for [<=50K,>50K] of each row of test data
proba_list=clf.predict_proba(np.array(X1_test.values))
proba_list
```

```python
#to obtain a dictionary of the class priors of output column
def getClassPriors(opcol):
    opcol=list(opcol)
    CP={i:opcol.count(i)/len(opcol) for i in np.unique(opcol)}
    return(CP)

getClassPriors(y_train['income'])
```

```python
# calculating the conditional probability of each category for each output value
def getcondProbs_obj(inp_traindf,op_traindf,opcol):
    Conpro={j:{k:(  len(inp_traindf[i][(inp_traindf[i]==j) & (op_traindf[opcol]==k)])  /  len(inp_traindf[i])  )   /   (  len(op_traindf[opcol][(op_traindf[opcol]==k)])  /  len(op_traindf[opcol])  )  for k in np.unique(op_traindf[opcol]) }  for i in inp_traindf.columns.values  for j in np.unique(inp_traindf[i]) }               
    return Conpro 

getcondProbs_obj(X_train,y_train,'income')       
```

```python
# predicting the output of a single row 
# for each possible output value, (multiplying together the condprob of each element of input row)
# and then declaring the greater value of multiplication as our predection

def predictone(x_obj,x_int,condprobs,classpriors):
    outprob={}
    for i in classpriors.keys():
        l=[condprobs[x_obj[j]][i] for j in x_obj.index]
        outprob[i]=np.prod(l)*classpriors[i]
    #print(outprob)  
    
    #creating a dictionary of gaussian nb predict_proba values
    k=clf.classes_
    v=clf.predict_proba([x_int])[0] 
    #remember, in clf.predict_proba([x_int]) prediction probabilities for outputs are ordered as their are output values in clf. classes_ 
    predvals=dict(zip(k,v))
    #print(predvals)
    
    #creating combined dictionary
    comb={k:predvals[k]*outprob[k] for k in clf.classes_ }
    #print(comb)
    
    return (max(comb, key=comb.get))

#clf.predict_proba([np.array(X1_test.values)[0]])[0]
predictone(X_test.iloc[0,0:7],X1_test.iloc[0,0:6],  getcondProbs_obj(X_train,y_train,'income'),  getClassPriors(y_train['income'])) 
```

```python
def predictall(X_test,X1_test,X2_test,condprobs,classpriors):
    y_pred=X2_test.apply(lambda row:predictone(row[X_test.columns],row[X1_test.columns],condprobs,classpriors), axis=1 )
    return y_pred

predictall(X_test, X1_test, X2_test,  getcondProbs_obj(X_train,y_train,'income'),  getClassPriors(y_train['income']))   
```

```python
def accuracy(actlist,predlist):
    acc=(sum([a==b for (a,b) in zip(predlist,actlist)])/len(predlist))*100
    print('Accuracy of Algorithm is : ',acc,'%')
    return acc
    
accuracy(y_test['income'].tolist(), predictall(X_test, X1_test, X2_test,  getcondProbs_obj(X_train,y_train,'income'),  getClassPriors(y_train['income'])))
```

Using K-fold cross validation (without shuffle)

(Using average of accuracies approach)

```python
import numpy as np
from sklearn.model_selection import KFold

#categorical data
X = dfobj.iloc[:,0:7]
y = dfobj.iloc[:,[-1]]

# X1 consists of only integer type columns, y1 consists of output column("income")
X1, y1 = dfint.iloc[:,:],df.iloc[:,[-1]]
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.3, random_state=5)

# X2 consists of all input columns, y2 consists of output column("income")
X2, y2 = df.iloc[:,0:13],df.iloc[:,[-1]]
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.3, random_state=5)

kf = KFold(n_splits=5)
#kf = KFold(n_splits=5, shuffle=True, random_state=5)
# by default shuffle is false, but we can use (shuffle=True) with a random state to split data after shuffling

kf.get_n_splits(X)

print(kf)
```

```python
acc=[]
for train_index, test_index in kf.split(X):
    print("TRAIN:", train_index, "TEST:", test_index)
    
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    X1_train, X1_test = X1.iloc[train_index], X1.iloc[test_index]
    y1_train, y1_test = y1.iloc[train_index], y1.iloc[test_index]
    
    X2_train, X2_test = X2.iloc[train_index], X2.iloc[test_index]
    y2_train, y2_test = y2.iloc[train_index], y2.iloc[test_index]
    
    acc.append(accuracy(y_test['income'].tolist(), predictall(X_test, X1_test, X2_test,  getcondProbs_obj(X_train,y_train,'income'),  getClassPriors(y_train['income']))))
    
print(acc)

#taking average of accuracies of all the folds
accavg=sum(acc)/len(acc)
accavg
```

Using append predictions to series approach (without shuffle)

This involves comparing all the predicted outputs at once with actual outputs instead of (comparing fold by fold and then taking average of accuracies). So, accuracy will be found out only once by appending predictions of all the folds together in a series and comparing it with the series of actual output values.

```python
pred_series=pd.Series()
for train_index, test_index in kf.split(X):
    print("TRAIN:", train_index, "TEST:", test_index)
    
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    X1_train, X1_test = X1.iloc[train_index], X1.iloc[test_index]
    y1_train, y1_test = y1.iloc[train_index], y1.iloc[test_index]
    
    X2_train, X2_test = X2.iloc[train_index], X2.iloc[test_index]
    y2_train, y2_test = y2.iloc[train_index], y2.iloc[test_index]
    
    pred_series=pred_series.append(predictall(X_test, X1_test, X2_test,  getcondProbs_obj(X_train,y_train,'income'),  getClassPriors(y_train['income'])))
    #we need to append and store the updated series in original pred_series, otherwise printing it would return an empty series

print(pred_series)

accuracy=(sum([a==b for (a,b) in zip(pred_series,df['income'])])/len(pred_series))*100
print('The final accuracy after appending all the predicted values in a series is : ',accuracy,'%')
```
